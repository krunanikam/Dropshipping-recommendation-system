```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#dataset
data = pd.read_csv('final_dataset.csv')
print(data.head())
```

```python
print(data.info())
```

```python
print(data.tail())
```

```python
df = pd.DataFrame(data)
plt.plot(df['Region'], df['Sales'], color='red', marker='o')
plt.title('Region vs Sales', fontsize=14)
plt.xlabel('Region', fontsize=14)
plt.ylabel('Sales', fontsize=14)
plt.grid(True)
plt.show()
```

```python
df = pd.DataFrame(data)
plt.plot(df['Profit'], df['Product_Category'], color='red', marker='o')
plt.title('Profit vs Product_Category', fontsize=14)
plt.xlabel('Profit', fontsize=14)
plt.ylabel('Product_Category', fontsize=14)
plt.grid(True)
plt.show()
```

```python
# reading the database
data = pd.read_csv("final_dataset.csv")
 
# Scatter plot with day against tip
plt.scatter(data['Sales'], data['Region'])
 
# Adding Title to the Plot
plt.title("Scatter Plot")
 
# Setting the X and Y labels
plt.xlabel('Sales')
plt.ylabel('Region')
 
plt.show()
```

```python
# reading the database
data = pd.read_csv("final_dataset.csv")
 
# Bar chart with day against tip
plt.bar(data['Product_Category'], data['Sales'])
 
plt.title("Bar Chart")
 
# Setting the X and Y labels
plt.xlabel('Product_Category')
plt.ylabel('Sales')
 
# Adding the legends
plt.show()
```

```python
# reading the database
data = pd.read_csv("final_dataset.csv")
 
# histogram of total_bills
plt.hist(data['Sales'])
 
plt.title("Histogram")
 
# Adding the legends
plt.show()
```
